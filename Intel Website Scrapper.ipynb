```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

PATH = "C:\chromedriver.exe"

driver=webdriver.Chrome(PATH)
driver.implicitly_wait(10)
driver.get("https://ark.intel.com/content/www/us/en/ark/products/series/202987/11th-generation-intel-core-i3-processors.html")

driver.set_window_size(1024, 600)
driver.maximize_window()


table_ele=driver.find_element_by_id("product-table")
headings=[]
table=[]
row=[]

tempList=table_ele.find_elements_by_tag_name("th")
for t in range(len(tempList)):
    if ((tempList[t].text).replace(" ","") != "" and tempList[t].text !="Compare\nAll | None"):
        headings.append(tempList[t].text)
table.append(headings)
tempList=[]
rows=table_ele.find_elements_by_tag_name("tr")
for count in range(len(rows)-1):
    tempList = rows[count].find_elements_by_xpath('//*[@id="product-table"]/tbody/tr['+str(count+1)+']/td')
    for t in range (len(tempList)):
        if(tempList[t].get_attribute("class")!="ark-compare compare-checkbox no-print" and tempList[t].get_attribute("style")!="display: none;"):
            data=tempList[t].text.replace(",","`").replace("\'","~")
            row.append(str(data))
    if row!=[]:
        table.append(row)
        row=[]
f=open('f1.txt','w')
for item in table:
    f.write("%s\n"%str(item).replace("[","").replace("]","").replace("\"","").replace("\'","").replace(",",";").replace("`",",").replace("~","\'"))
f.close()
```
